```
import numpy as np
from PIL import Image
import pandas as pd

train_dir_list = os.listdir(unzip_train_path)
test_dir_list = os.listdir(unzip_test_path)
train_dir_path = unzip_train_path
test_dir_path = unzip_test_path
temp = np.empty((960 * 540, 1))

tmp_path = []
for subdir in train_dir_list:
    tmp_path.append(os.path.join(train_dir_path, subdir))

print('len(tmp_path) = ', len(tmp_path))

def get_imgpath(path_list):
    tmp_list = []
    for path in path_list:
        subdir = os.listdir(path)
        for dir in subdir:
            result = os.path.join(path, dir)
            tmp_list.append(result)
    print('len(return_list) = ', len(tmp_list))
    return tmp_list

train_img_path_list = get_imgpath(tmp_path)
train_img_list = get_imgpath(train_img_path_list)

tmp_path = []
for subdir in test_dir_list:
    tmp_path.append(os.path.join(test_dir_path, subdir))

print('len(tmp_path) = ', len(tmp_path))

test_img_list = get_imgpath(tmp_path)

def jpg_to_numpy(path_list):
    for path in path_list:
        try:
            img = Image.open(path)
            img_to_array = np.array(img).reshape(-1, 1)
            array = np.append(np.empty((960 * 540, 1)), img_to_array, axis=0)
        except Exception as err:
            print(str(err))
    return array

train_data = jpg_to_numpy(train_img_list)
test_data = jpg_to_numpy(test_img_list)

print('train_data.shape = ', train_data.shape, ', train_data.ndim = ', train_data.ndim)
print('test_data.shape = ', test_data.shape, ', test_data.ndim = ', test_data.ndim)
```
