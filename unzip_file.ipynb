```
# unzip file

import zipfile
import os


def unzip(dir_path, zip_file, unzip_folder=None):
    if unzip_folder == None:
        unzip_path = dir_path
    else:
        unzip_path = os.path.join(dir_path, unzip_folder)
    print('unzip path:', unzip_path)

    with zipfile.ZipFile(os.path.join(dir_path, zip_file)) as f:
        name_list = sorted(f.namelist())
        print('name_list: ', name_list[:5])
        for name in name_list:
            try:
                f.extract(name, unzip_path)
            except Exception:
                print('Excepted file name: ', name)
    print('len(os.listdir(unzip_path)) = ', len(os.listdir(unzip_path)))

default_path = '/content/gdrive/MyDrive/data'
dir_path = os.path.join(default_path, 'dirty mnist')
print('dir_path: ', dir_path)
```
